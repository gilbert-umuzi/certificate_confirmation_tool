```python
import requests
import csv
import os
import time
import random
from google.colab import userdata
```

```python
# Your Airtable settings
API_KEY = userdata.get('Airtable_API_Key') # Stored in Colab keys
BASE_ID = 'apppTFgbfEg49n29R'
TABLE_NAME = 'SAP Portal Qualifiers'
VIEW_NAME = 'Boost Your Career completers'
```

```python
# Certificate Names (Columns containing attachments)
certificate_names = [
    'Boost Your Career with SAP Skills My Learning screenshot',
    'Learn how to learn (Google)',
    'Business Communication (Google)',
    'Communicate your ideas through storytelling and design (Google)',
    'Get started with Microsoft Teams',
    'Tech for Good: The Role of ICT in Achieving the SDGs'
]
```

```python
# Function to convert seconds to hours, minutes, and seconds
def convert_seconds_to_time(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return hours, minutes, seconds
```

```python
# Download view as CSV

url = f"https://api.airtable.com/v0/{BASE_ID}/{TABLE_NAME}"
params = {'view': VIEW_NAME}
headers = {'Authorization': f'Bearer {API_KEY}'}

fieldnames = set()
records_list = []

# Collect all records and dynamically update fieldnames
while True:
    response = requests.get(url, headers=headers, params=params)
    records = response.json().get('records')

    for record in records:
        records_list.append(record['fields'])
        for key in record['fields'].keys():
            fieldnames.add(key)

    offset = response.json().get('offset')
    if offset is None:
        break

    params['offset'] = offset

# Write to CSV
with open('view_2.csv', 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

    for record in records_list:
        csv_writer.writerow(record)
```

```python
# Download CSV and attachments

# Endpoint URL
url = f"https://api.airtable.com/v0/{BASE_ID}/{TABLE_NAME}"

# Headers
headers = {
    'Authorization': f'Bearer {API_KEY}'
}

# Parameters (Filter by View and Limit Records for Sample)
params = {
    'view': VIEW_NAME,
    'maxRecords': 1000  # Control sample size here
}

# Make the request
start_time = time.time()
response = requests.get(url, headers=headers, params=params)
airtable_records = response.json().get('records')
end_time = time.time()

# Estimate time
sample_time = end_time - start_time
total_estimate = (sample_time / 10) * len(airtable_records)
print(f"Estimated Time: {total_estimate} seconds for {len(airtable_records)} records")

# Prepare for CSV export
csv_file = 'exported_view.csv'
csv_data = []

for record in airtable_records:
    fields = record.get('fields')
    csv_data.append(fields)

# Write to CSV
keys = csv_data[0].keys()
with open(csv_file, 'w', newline='') as output_file:
    dict_writer = csv.DictWriter(output_file, fieldnames=keys)
    dict_writer.writeheader()
    dict_writer.writerows(csv_data)

# Download attachments
for record in airtable_records:
    fields = record.get('fields')

    for certificate_name in certificate_names:
        attachments = fields.get(certificate_name)

        if attachments:
            folder_path = f"/content/drive/MyDrive/Colab Notebooks/Certificate confirmer data/attachments/{certificate_name}"

            if not os.path.exists(folder_path):
                os.makedirs(folder_path)

            for attachment in attachments:
                attachment_url = attachment.get('url')
                attachment_name = attachment.get('filename')

                if os.path.exists(f"{folder_path}/{attachment_name}"):
                    print(f"Skipping {attachment_name}, already exists.")
                    continue

                print(f"Downloading {attachment_name} ...")
                r = requests.get(attachment_url)
                with open(f"{folder_path}/{attachment_name}", 'wb') as f:
                    f.write(r.content)
                print(f"{attachment_name} downloaded.")
```

```python
# Count attachments downloaded
import os

def count_files_in_directory(directory):
    total_files = 0

    for root, dirs, files in os.walk(directory):
        total_files += len(files)

    return total_files

directory = '/content/attachments'  # Replace with the actual path
total_downloaded_attachments = count_files_in_directory(directory)

print(f"Total downloaded attachments: {total_downloaded_attachments}")
```
